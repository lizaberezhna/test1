```python
# –ö–ª—ñ—Ç–∏–Ω–∫–∞ 2: –ü–µ—Ä–µ–≤—ñ—Ä–∫–∞ –Ω–∞—è–≤–Ω–æ—Å—Ç—ñ —Ñ–∞–π–ª—ñ–≤ –º–æ–¥–µ–ª—ñ
import os
import glob

print("üîç –ü–æ—à—É–∫ –Ω–∞–≤—á–µ–Ω–æ—ó –º–æ–¥–µ–ª—ñ...")

# –ü—Ä–æ—Å—Ç–∏–π –ø–æ—à—É–∫
model_files = []
for ext in ['.h5', '.keras', '.hdf5']:
    for path in glob.glob(f'**/*{ext}', recursive=True):
        if os.path.getsize(path) > 10000:  # –ë—ñ–ª—å—à–µ 10KB
            model_files.append(path)

if model_files:
    print(f"‚úÖ –ó–Ω–∞–π–¥–µ–Ω–æ {len(model_files)} —Ñ–∞–π–ª(—ñ–≤) –º–æ–¥–µ–ª—ñ:")
    for i, path in enumerate(model_files, 1):
        size_kb = os.path.getsize(path) / 1024
        print(f"{i}. {path} ({size_kb:.1f} KB)")

    # –ê–≤—Ç–æ–º–∞—Ç–∏—á–Ω–æ –≤–∏–±–∏—Ä–∞—î–º–æ –ø–µ—Ä—à–∏–π
    MODEL_PATH = model_files[0]
    print(f"\nüéØ –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ: {MODEL_PATH}")
else:
    print("‚ùå –ú–æ–¥–µ–ª—å –Ω–µ –∑–Ω–∞–π–¥–µ–Ω–∞!")
    print("\nüìã –í–∏–±–µ—Ä—ñ—Ç—å –≤–∞—Ä—ñ–∞–Ω—Ç:")
    print("1. –°—Ç–≤–æ—Ä–∏—Ç–∏ –ø—Ä–æ—Å—Ç—É —Ç–µ—Å—Ç–æ–≤—É –º–æ–¥–µ–ª—å")
    print("2. –í–∫–∞–∑–∞—Ç–∏ —à–ª—è—Ö –≤—Ä—É—á–Ω—É")
    print("3. –í–∏–π—Ç–∏")

    choice = input("\n–í–∞—à –≤–∏–±—ñ—Ä (1/2/3): ").strip()

    if choice == '1':
        print("\n–°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–µ—Å—Ç–æ–≤–æ—ó –º–æ–¥–µ–ª—ñ...")
        import tensorflow as tf
        import numpy as np

        # –ü—Ä–æ—Å—Ç–∞ –º–æ–¥–µ–ª—å
        model = tf.keras.Sequential([
            tf.keras.layers.Dense(64, activation='relu', input_shape=(8,)),
            tf.keras.layers.Dense(32, activation='relu'),
            tf.keras.layers.Dense(16, activation='relu'),
            tf.keras.layers.Dense(1)
        ])

        model.compile(optimizer='adam', loss='mse')
        model.save('test_model.h5')
        MODEL_PATH = 'test_model.h5'
        print(f"‚úÖ –°—Ç–≤–æ—Ä–µ–Ω–æ: {MODEL_PATH}")

    elif choice == '2':
        user_path = input("–í–≤–µ–¥—ñ—Ç—å –ø–æ–≤–Ω–∏–π —à–ª—è—Ö –¥–æ –º–æ–¥–µ–ª—ñ: ").strip()
        if os.path.exists(user_path):
            MODEL_PATH = user_path
            print(f"‚úÖ –í–∏–∫–æ—Ä–∏—Å—Ç–æ–≤—É—î–º–æ: {MODEL_PATH}")
        else:
            print("‚ùå –§–∞–π–ª –Ω–µ —ñ—Å–Ω—É—î!")
            exit()
    else:
        print("‚ùå –°–∫–∞—Å–æ–≤–∞–Ω–æ.")
        exit()
```

```python
# –ö–ª—ñ—Ç–∏–Ω–∫–∞ 1: –°—Ç–≤–æ—Ä–µ–Ω–Ω—è —Ç–∞ –∑–±–µ—Ä–µ–∂–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ
import tensorflow as tf
import numpy as np
import os

print("=" * 60)
print("ü§ñ –°–¢–í–û–†–ï–ù–ù–Ø –¢–ê –ö–û–ù–í–ï–†–¢–ê–¶–Ü–Ø –ú–û–î–ï–õ–Ü")
print("=" * 60)

# 1. –°—Ç–≤–æ—Ä—é—î–º–æ –ø—Ä–æ—Å—Ç—É –º–æ–¥–µ–ª—å
print("–°—Ç–≤–æ—Ä–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ...")
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(8,)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1)
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
print("‚úÖ –ú–æ–¥–µ–ª—å —Å—Ç–≤–æ—Ä–µ–Ω–∞")

# 2. –¢—Ä–µ–Ω—É—î–º–æ –Ω–∞ —Å–∏–Ω—Ç–µ—Ç–∏—á–Ω–∏—Ö –¥–∞–Ω–∏—Ö
print("–¢—Ä–µ–Ω—É–≤–∞–Ω–Ω—è –Ω–∞ —Å–∏–Ω—Ç–µ—Ç–∏—á–Ω–∏—Ö –¥–∞–Ω–∏—Ö...")
np.random.seed(42)
X_train = np.random.randn(1000, 8).astype(np.float32)
y_train = np.random.randn(1000, 1).astype(np.float32)

model.fit(X_train, y_train, epochs=10, verbose=0, batch_size=32)
print("‚úÖ –ú–æ–¥–µ–ª—å –Ω–∞–≤—á–µ–Ω–∞")

# 3. –ó–±–µ—Ä—ñ–≥–∞—î–º–æ –≤ –¥–µ–∫—ñ–ª—å–∫–æ—Ö —Ñ–æ—Ä–º–∞—Ç–∞—Ö
print("–ó–±–µ—Ä–µ–∂–µ–Ω–Ω—è –º–æ–¥–µ–ª—ñ...")
model.save('housing_model_simple.h5')
print("‚úÖ –ó–±–µ—Ä–µ–∂–µ–Ω–æ —è–∫ housing_model_simple.h5")

# –¢–µ—Å—Ç –ø—Ä–æ–≥–Ω–æ–∑—É–≤–∞–Ω–Ω—è
test_input = np.array([[3.87, 28.6, 5.43, 1.10, 1425, 3.10, 35.63, -119.57]])
prediction = model.predict(test_input, verbose=0)
print(f"üìä –¢–µ—Å—Ç–æ–≤–∏–π –ø—Ä–æ–≥–Ω–æ–∑: {prediction[0][0]:.4f}")
print(f"üíµ –£ –¥–æ–ª–∞—Ä–∞—Ö: ${prediction[0][0] * 100000:,.0f}")
```
