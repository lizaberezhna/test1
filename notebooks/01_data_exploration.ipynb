```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Завантаження даних
california = fetch_california_housing()
data = pd.DataFrame(california.data, columns=california.feature_names)
data['Price'] = california.target

print("Розмір датасету:", data.shape)
print("\nПерші 5 рядків:")
print(data.head())

print("\nОсновна статистика:")
print(data.describe())

print("\nІнформація про типи даних:")
print(data.info())

print("\nПеревірка пропущених значень:")
print(data.isnull().sum())

# Візуалізація
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(data.columns[:-1]):
    axes[i].scatter(data[col], data['Price'], alpha=0.5)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Price')
    axes[i].set_title(f'{col} vs Price')

plt.tight_layout()
plt.show()

# Кореляційна матриця
plt.figure(figsize=(10, 8))
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Кореляційна матриця')
plt.tight_layout()
plt.show()

# Гістограми
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

for i, col in enumerate(data.columns):
    axes[i].hist(data[col], bins=30, edgecolor='black', alpha=0.7)
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Частота')
    axes[i].set_title(f'Розподіл {col}')

plt.tight_layout()
plt.show()
```
